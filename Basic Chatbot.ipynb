```python
import nltk
from nltk.chat.util import Chat, reflections

# Define patterns and responses
patterns_responses = [
    (r'hi|hello|hey', ['Hello! How can I help you today?', 'Hi there! What can I do for you?', 'Hey! How can I assist you?']),
    (r'what is your name\??', ['I am a chatbot created to help you.', 'You can call me ChatBot.', 'My name is ChatBot, and I am here to assist you.']),
    (r'how are you\??', ['I am just a computer program, but thanks for asking!', 'I am here to assist you. How can I help?', 'I am doing well, thank you! How can I assist you today?']),
    (r'quit', ['Bye! Have a great day!', 'Goodbye!', 'Take care! See you soon.']),
    (r'what can you do\??', ['I can assist you with your questions and provide information.', 'I am here to help with whatever you need.', 'I can answer your questions, provide information, and assist with various tasks.']),
    (r'what time is it\??', ['I don’t have a watch, but you can check the time on your device!', 'I can’t tell the time, but your device should be able to.', 'Please check the time on your device.']),
    (r'thank you|thanks', ['You’re welcome!', 'Happy to help!', 'No problem!']),
    (r'who created you\??', ['I was created by a team of developers.', 'A team of programmers designed me to assist you.', 'I was developed by software engineers.']),
    (r'what is python\??', ['Python is a popular programming language known for its simplicity and readability.', 'Python is used for web development, data analysis, AI, and more.', 'Python is a versatile programming language that is easy to learn.']),
    (r'can you help me with (.*)\??', ['Sure, I’d be happy to help with {0}!', 'Of course! What do you need to know about {0}?', 'I can assist you with {0}. Let’s get started!']),
    (r'(.*) your favorite (.*)\??', ['I don’t have personal preferences, but I can help you find information about {1}.', 'As a bot, I don’t have favorites, but I can provide information on {1}.']),
    (r'(.*)', ['Sorry, I do not understand that.', 'Can you please rephrase your question?', 'I’m not sure I understand. Can you clarify?'])
]

def create_chatbot():
    # Create a Chat object with the patterns and responses
    chatbot = Chat(patterns_responses, reflections)
    return chatbot

def chat_with_user(chatbot):
    print("Hello! I am your chatbot. Type 'quit' to exit.")

    while True:
        user_input = input("You: ")
        response = chatbot.respond(user_input)
        print(f"ChatBot: {response}")

        if user_input.lower() == 'quit':
            break

if __name__ == "__main__":
    chatbot = create_chatbot()
    chat_with_user(chatbot)
```
